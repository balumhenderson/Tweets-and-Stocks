```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import datetime
```

```python
trump_tweets = pd.read_csv("TrumpTweets.csv")
musk_tweets = pd.read_csv("MuskTweets.csv")
west_tweets = pd.read_csv("WestTweets.csv")
biden_tweets = pd.read_csv("BidenTweets.csv")
stocks_norm = pd.read_csv("StockData_Normalised.csv")
```

```python
fig = make_subplots(specs=[[{"secondary_y":True}]])

fig.add_trace(
    go.Scatter(
        x=stocks_norm["Unnamed: 0"],
        y=stocks_norm["NASDAQ"],
        name="NASDAQ",
        mode="lines"),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(
        x=stocks_norm["Unnamed: 0"],
        y=stocks_norm["USD-GBP"],
        name="USD-GBP",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(
        x=stocks_norm["Unnamed: 0"],
        y=stocks_norm["USD-JPY"],
        name="USD-JPY",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(
        x=stocks_norm["Unnamed: 0"],
        y=stocks_norm["USD-CAD"],
        name="USD-CAD",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(
        x=stocks_norm["Unnamed: 0"],
        y=stocks_norm["Bitcoin"],
        name="Bitcoin",
        mode="lines",
        visible=True,
    ),
    secondary_y=False,
)
fig.add_trace(
    go.Scatter(
        x=trump_tweets["Date/Time"],
        y=trump_tweets["Retweet Rolling Av"],
        name="Trump Retweets",
        mode="lines",
        visible=True,
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=trump_tweets["Date/Time"],
        y=trump_tweets["Favourite Rolling Av"],
        name="Trump Favourites",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=musk_tweets["Date/Time"],
        y=musk_tweets["Retweet Rolling Av"],
        name="Musk Retweets",
        mode="lines",
        visible=True,
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=musk_tweets["Date/Time"],
        y=musk_tweets["Favourite Rolling Av"],
        name="Musk Favourites",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=west_tweets["Date/Time"],
        y=west_tweets["Retweet Rolling Av"],
        name="West Retweets",
        mode="lines",
        visible=True,
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=west_tweets["Date/Time"],
        y=west_tweets["Favourite Rolling Av"],
        name="West Favourites",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=biden_tweets["Date/Time"],
        y=biden_tweets["Retweet Rolling Av"],
        name="Biden Retweets",
        mode="lines",
        visible=True,
    ),
    secondary_y=True,
)
fig.add_trace(
    go.Scatter(
        x=biden_tweets["Date/Time"],
        y=biden_tweets["Favourite Rolling Av"],
        name="Biden Favourites",
        mode="lines",
        visible="legendonly",
    ),
    secondary_y=True,
)



# Add figure title
fig.update_layout(
    title_text="Tweets and Stocks",
    showlegend=True,
    plot_bgcolor="rgb(240, 260, 250)",
    xaxis_range=[datetime.datetime(2018, 1, 1),
                datetime.date.today()]
)

# Set x-axis title
fig.update_xaxes(
    title = "Date/Time",
)

# Set y-axes titles
fig.update_yaxes(title_text="Normalised Stock Performance", secondary_y=False)
fig.update_yaxes(title_text="Tweet Retweets/Favourites", secondary_y=True)

fig.show()
```
